```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r'/Users/Downloads/advertising.csv'
df_ad_data = pd.read_csv(file_path)
df_ad_data.head()
```

# Radio

```python
X_radio = df_ad_data[["Radio"]]
X_radio.head()
```

```python
y_radio = df_ad_data['Sales']
y_radio.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train_radio, X_test_radio, y_train_radio, y_test_radio = train_test_split(X_radio, y_radio, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train_radio, y_train_radio)
```

```python
test_predictions_radio = model.predict(X_test_radio)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(y_test_radio, test_predictions_radio)
```

```python
np.sqrt(mean_squared_error(y_test_radio, test_predictions_radio))
```

```python
test_residuals_radio = y_test_radio - test_predictions_radio
```

```python
sns.scatterplot(x=y_test_radio, y=test_residuals_radio)
plt.axhline(y=0, color='r', ls='--')
```

```python
sns.displot(test_residuals_radio, bins=25, kde=True)
```
